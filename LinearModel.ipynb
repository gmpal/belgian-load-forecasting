```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from utils import preprocess, create_time_series_splits, mape
from sklearn.linear_model import LinearRegression
```

```python
df = preprocess("./ods001.csv")
```

```python
excluded_columns = ['Total Load', 'Most recent forecast', 'Most recent P10',
       'Most recent P90', 'Day-ahead 6PM forecast', 'Day-ahead 6PM P10',
       'Day-ahead 6PM P90', 'Week-ahead forecast', 'Year', 'Month', 'Day',
       'Hour', 'Minute', 'Season', 'Lockdown']
```

```python
splits = create_time_series_splits(
    data=df,
    train_size_days=5,
    test_size_days=2,
    num_splits=3,
    window_size_steps=3,
    exclude_columns=excluded_columns,
    target_column='Total Load Interpolated',
    prediction_horizon_steps=4,
    shifting_steps=4,
    elia_column_to_return='Most recent forecast'
)
```

First, we assess the quality of ELIA's forecasting, in terms of MAPE:

```python
first_split = splits[0]
X_train = first_split['X_train']
Y_train = first_split['Y_train']
X_test = first_split['X_test']
Y_test = first_split['Y_test']
ELIA_train = first_split['ELIA_train']
ELIA_test = first_split['ELIA_test']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Y_test["Total Load Interpolated_t+1"], label="Total Load")
plt.plot(ELIA_test["Most recent forecast_t+1"], label="Most recent forecast")
plt.title(f"ELIA Forecasting example\nMAPE: {mape(ELIA_test['Most recent forecast_t+1'], Y_test['Total Load Interpolated_t+1']):.4f}%")
plt.legend()
plt.grid()
plt.show()
```

We show that a simple linear model outperforms ELIA's forecasts:

```python
lm = LinearRegression().fit(X_train, Y_train)
y_hat = lm.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Y_test["Total Load Interpolated_t+1"].to_numpy(), label="Total Load")
plt.plot(y_hat[:, 0], label="Linear Model forecast")
plt.title(f"Linear Model Forecasting example\nMAPE: {mape(y_hat[:, 0], Y_test['Total Load Interpolated_t+1']):.4f}%")
plt.legend()
plt.grid()
plt.show()
```
